# Data Preprocessing

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast 
# .)the blw fun will convert th string list to list . 
# ast.literal_eval()
```

```python
# .)Preprocesing on genres  and keywords column column . 

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# .)Preprocesing on cast column column . 

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name']) 
            counter+=1
        else:
            break
    return L 
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
# .)Preprocesing on crew column column . 
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name']) 
            break
    return L 
```

```python
movies['crew']= movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python

movies['overview'][0]
```

```python
# overview column is string now i am converting it intolist . 
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# removing the spaces between the word of column genres,keywords cast
# and crew columns . 
movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# makin new columns tags and i will concat the genres,keywords cast and crew columns. 
new_df = movies[['movie_id','title','tags']]
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# Apply stemming technique on above words . 

# befor stemming
['loved','loving','love']

# after stemming
['love','love','love']
```

```python
# stemming library is nltk.
import nltk
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
ps.stem('loved')
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
similarity = cosine_similarity(vectors)
```

```python
# makin recmommnd fun it will retun the name of top 5 similar movies from 
# with cosine simlarity . 
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
